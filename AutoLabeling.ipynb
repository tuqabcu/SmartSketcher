```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
import os
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp17/weights/last.pt', force_reload=True)
```

```python
model.conf = 0.7
for path, subdirs, files in os.walk("yolov5/data/sub_dataset"):
    for name in files:
        img = path + "/" + name
        print(name)
        results = model(img)
        print(results.xyxyn[0][:, :].numpy())
```
